## Formulation of problem
* Elliptic problem with incomplete boundary conditions

\begin{equation*}
- \Delta u = f,\quad x\in {\Omega}, \quad u = g,\quad, x\in \Gamma = \partial{\Omega}
\end{equation*}
The source term $f$ is known and $g$ is unknown and is some function in $K^B$, ball of a space that compactly imported in $H^{\frac12}(\Gamma)$
One typical choice is the unit ball $U(H^s(\Gamma))$ in $H^s(\Gamma)$. 

### Harmonic recovery 
Introducing the space of harmonic functions $\mathcal{H}^s({\Omega})$, which is also the extension of $H^s(\Gamma)$ by harmonic extension.
The norm of $\mathcal{H}^s({\Omega} )$ is defined by the boundary norm.
According to the smoothness of the domain boundary, we have the interpolation
\begin{equation*}
\mathcal{H}^s({\Omega})\subset H^r({\Omega})
\end{equation*}
with $r=\min\{s+\frac12,r^*\}$

* For smooth domain, $r^*=\infty$
* For polygons, $r^* = 1+ \frac{\pi}{\omega}-\varepsilon$ where $\omega$ is the largest inner angle. 

Example: $\omega = \frac{\pi}{2}$, which is convex, then $r^*\approx 2$

The task is to find the Chebyshev center in ${H}^1$ which satisfies the observations (in a linear space)
\begin{equation*}
\lambda_j( \phi) = w_j
\end{equation*}
The candidate set is denoted as $\mathcal{K}_w = \{u\in \mathcal{H}^s: \lambda(u) = w\}$.

First in the hilbert space $\mathcal{H}^s$, the Chebyshev center coincides with the minimal norm interpolant
\begin{equation*}
u^*(w) = \textrm{argmin} \{ \| v\|_{\mathcal{H}^s}: v\in \mathcal{H}^s, \ \lambda(v) = w\} 
\end{equation*}

Due to the symmetry of $\mathcal{K}_w$, its Chebyshev center in $H^1$ is its geometrical center and the Chebyshev center in $\mathcal{H}^s$

Finally, to solve the minimal norm interpolant, we apply the $L^2$ projection in the Hilbert space, where the finite dimensional space is spanned by the Riesz representers of functionals $\lambda$

## Computation of Riesz representer in Harmonic space

* Exact riesz representer

\begin{align*}
& \phi = E \psi,\\
& \langle \psi, \eta \rangle_{H^s(\Gamma)} = \lambda(E\eta),\quad \forall \eta\in H^s(\Gamma)
\end{align*}
which means solving boundary representer first and exnteded to the domain

* Finite element setting 

Consider a finite element space $V_h$ associated with a family of meshes $\mathcal{T}_h$ of ${\Omega}$.
The boundary space $T_h = T(V_h)$.
Domain space with Dirichlet boundary condition $W_h = \{v_h\in V_h: T(v_h) = 0\}$.

* The constrain of $P_k$ Lagrange element on the boundary belongs to $H^{\frac32-}(\Gamma)$. Then $T_h\subset H^s(\Gamma)$ for $s<\frac32$.
* Numerical Harmonic extension $E_h: T_h\to V_h$

\begin{equation*}
\int_ {\Omega} \nabla E_h g_h\cdot \nabla v_h = 0,\quad v_h \in W_h
\end{equation*}
* Riesz representer

\begin{align*}
& \phi_h = E_h \psi_h,\\
& \langle \psi_h,g_h \rangle_{H^s(\Gamma)} = \lambda(E_h g_h),\quad g_h\in T_h
\end{align*}

* Saddle point formulation

Introduce $X^s= \{v\in H^1( {\Omega}): v_ \Gamma \in H^s( \Gamma)\}$ (bigger than $H^1({\Omega})\cap \mathcal{H}^s({\Omega}))$, equipped with the norm
\begin{equation*}
\| v\|_{X^s({\Omega})} = \left( \| v_ \Gamma\|_{H^s(\Gamma)}^2 + \| \nabla v\|_{L^2({\Omega})}^2\right)^{\frac12}
\end{equation*}
Note that $\| v\|_{\mathcal{H}^s} = \| v\|_{H^s(\Gamma)}$ controls the $L^2$ norm

The saddle point problem formulates as find $(\phi,\pi)\in X^s({\Omega})\times H^1({\Omega})$ such that 
\begin{align*}
& a(\phi,v) + b(v,\pi) = \lambda(v),\quad v\in X^s({\Omega})\\
& b(\phi,z) = 0, z\in H_0^1({\Omega})
\end{align*}
where 
\begin{align*}
a(\phi,v) = \langle \phi_ \Gamma, v_ \Gamma \rangle_{H^s(\Gamma)},\quad
b(v, \pi) = \langle \nabla v, \nabla \pi \rangle_{L^2({\Omega})}
\end{align*}
Testing $v\in \mathcal{H}^s({\Omega})$ in the first equation leads to $a( \phi,v) = \lambda(v)$
* Continuity of $a$ and $b$; Coercivity of $a$ on the null space of $b$ in $X^s({\Omega})$, null space is $\mathcal{H}^s({\Omega})$

For $v\in \mathcal{H}^s({\Omega})$, one has
\begin{equation*}
\| v\|_{X^s({\Omega})}^s \le \| v_ \Gamma\|_{H^s(\Gamma)}^2 + \| v\|_{H^1}^2 \le (1+C^2) \| v_ \Gamma\|_{H^s(\Gamma)}^2\le (1+C^2)a(v,v)
\end{equation*}
by using the embedding $H^s(\Gamma)\to H^\frac12(\Gamma)$


## Stability and convergence
* Stability of $E$
Since $E$ is to solve the Laplace equation, we obtain the stability 
\begin{equation*}
\| Eg\|_{H^1({\Omega})} \lesssim \| g\|_{H^\frac12(\Gamma)}
\end{equation*}
* Stability of $E_h$
If there exists uniformly $H^1$ stable projections $P_h: H^1({\Omega})\to V_h$ that preserves the homogeneous boundary condition (to apply Poincare inequality and $\|\nabla E_h g_h\|\le \|\nabla P_h Eg_h\|$)
\begin{equation*}
P_h(H_0^1({\Omega})) = W_h,\quad \| P_hv\|_{H^1}\le B \| v\|_{H^1}
\end{equation*}
where $B$ is independent of $h$.
This guarantees the stability of $E_h:T_h\to V_h$
\begin{align*}
\| E_h g_h - P_hEg_h\|_{H^1}\le 2C_P \| P_h Eg_h\|_{H^1}
\end{align*}
* Error in replacing $E$ by $E_h$.
For any $g_h\in T_h$, one has
\begin{align*}
\| (E-E_h) g_h\|_{H^1({\Omega})} \le C(r,s, \sigma, {\Omega}) h^{r-1} \| g_h\|_{H^s(\Gamma)}
\end{align*}
* Regularity of boundary Riesz representer
\begin{equation*}
\langle \kappa, v \rangle_{H^s(\Gamma)} = \lambda(v),\quad v\in H^s(\Gamma)
\end{equation*}

    If $\lambda\in H^{-s}(\Gamma)$, then $\kappa \in H^s(\Gamma)$.

    If $\lambda\in L^2(\Gamma)$, then $\kappa\in H^1(\Gamma)$ and $\kappa \in H^2(E)$ for each edge $E$ such that the 
standard finite element approximation estimes yield 
\begin{equation*}
\min_{\kappa_h\in T_h} \| \kappa - \kappa_h\|_{H^1(\Gamma)}\le Ch \| g\|_{L^2(\Gamma)}
\end{equation*}

* More on regularity

    Consider the two point boundary value problem with discontinuous $\kappa$ such that 
    
\begin{equation*}
\kappa = 1_{[0,0.5)} +  2*1_{[0.5,1]}
\end{equation*}
    The weak form writes 
    
\begin{equation*}
\int_0^1 \kappa u^\prime v^\prime = \int_0^1 f v
\end{equation*}
for $v\in H_0^1$

A typical choice is choosing $f = -\frac14x^{-\frac54}$, as the derivative of $g = x^{-\frac14}$.
By Fourier transform, $x^{-\alpha}\to \xi^{-(n-\alpha)}$, then ${\mathcal{F}} g = \xi^{-\frac34}$ and allows to take $\frac14$ derivatives
Means $g\in H^{\frac14-}$.
Then $f\in H^{-\frac34-}$.
Then $u\in H^{\frac14-}$ (no elliptic regularity improvement, only valid in two sub-intervals)
Then linear approximation error is $h^{\frac14}$.
